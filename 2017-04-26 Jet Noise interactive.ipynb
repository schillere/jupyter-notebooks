## 11.7 Jet Noise

(page 444)

Pneumatic devices ejecting gas in the form of high-pressure jets can be significant generators of noise...

The acoustical power generated by subsonic jet in free space is related to the mechanical stream power by an efficiency factor as follows:
\begin{equation*}
W_{a}=\eta W_{m}
\quad\mathrm{(W)}
\tag{11.21}
\end{equation*}

Stream mechanical power, $W_{m}$ is equal to the convected kinetic energy of the stream, whihc for a jet of circular cross-section is:

\begin{equation*}
W_{m}=\rho U^{3} \pi d^{2} / 8 
\quad\mathrm{(W)}
\tag{11.22}
\end{equation*}

where: $U$ is the jet exit velocity (m/s), $W_{a}$ is the mechanical stream power (W), $\eta$ is the acoustic efficiency of the jet, $d$ is the jet diameter (m), and $\rho$ is the density of the gas ($kg/m^{3})$.

The acoustic efficiency of a jet is approximately:

$$
\eta=(T/T_{0})^{2}(\rho/\rho_{0})K_{a}M^{5}
\tag{11.23}
$$

for the range of $ 0.1 \geq M \leq 1.0$, and for the range of the quantity $(T/T_{0})^{2}(\rho/\rho_{0})$ between 0.1 and 10. In equation 11.23, $\rho_{0}$ is the density $(kg/m^{3})$ of the ambient gas, $K_{a}$ is the acoustical poweqr coefficient and is approximately $5\times10^{-5}$, $M$ is the stream Mach number relative to the ambient gas, $T$ is the jet absolute temperature (K), and $T_{0}$ is the absolute temperature (K) of the ambient gas.

If the jet pressure ratio is less than 1.89, hte jet is subsonic, and Equation 11.23 may be used to estimate $\eta$. if the pressure ratio is greater than 1.89, the gas exit velocity will equal Mach 1. If the nozzle contractraction is followed by an expansion, the gas velocity may exceed Mach 1.

The overall sound power level of the jet is:
$$
L_{w}=10log_{10}W_{a}+120
\quad \mathrm{(dB\,re\,}10^{-12} \mathrm{W)}
\tag{11.24}
$$

In a free field or close to the jet, the overall sound pressure level is given by:

$$
L_{p}=L_{w}+DI-10log_{10}(4\pi r^{2})
\quad\mathrm{(dB\,re\,20\,\mu Pa)}
\tag{11.25}
$$
where $DI$ is the jet directivity index and $r$ is the distance (m) fromthe jet orifice to the observation point.

### Example 11.3
Gas is to be exhausted under pressure from a 100 mm diameter pipe to the atmosphere at a speed of 230 m/s. The density of the gas is 1.5 kg/m^3 and its temperature is that of the ambient atmosphere into which it is to be exhausted. Determine the overall sound pressure levels at 10 m from the jet orifice and at various angles relative to the direction of the jet.

1) Determine the jet efficiency using Equation 11.23. First calculate $(T/T_{0})^{2}(\rho/\rho_{0})$

```python
#declare symbols
var('W_a eta W_m');
var('W_m rho U pi d');
var('T T_0 rho rho_0 K_a M');
var('L_w L_p DI r');
var('c c_0');
```

```python
W_a = eta * W_m
W_m = rho * U**3 * pi * (d**2)/8
eta = ((T/T_0)**2) * (rho/rho_0) * K_a * M**5
L_w = 10*log(W_a,10)+120
L_p = L_w + DI - 10*log((4*pi*r**2),10)
```

```python
#Values in preferred units
_U = 230 #m/s, jet exit velocity
_rho = 1.5 #kg/m^3, density of jet gas
_rho_0 = 1.1295 #kg/m^3, density of ambient gas
_T = 77 #F, absolute temperature of jet gas
_T_0 = 77 #F, absolute temperature of the ambient gas
_c_0 = 344 #m/s, speed of sound in air surrounding jet
_K_a = 5*10**-5 #acoustical power coefficient
_d = 0.1 #m, jet diameter
_r = 10 #m, distance from jet orifice to observation point
_DI = 8 #dB, directional correction for jets

#correct units for equation
_T = funFtoK(_T) #K
_T_0 = funFtoK(_T_0) #K

subList = {U:_U, rho:_rho, rho_0:_rho_0, T:_T, T_0:_T_0, c_0:_c_0, K_a:_K_a, pi:math.pi, d:_d, r:_r, DI:_DI}
```

```python
(((T/T_0)**2) * (rho/rho_0)).subs(subList)
```

The speed of sound is assumed to be 344 m/s in the air surrounding the jet...

```python
M = U/c_0 #Mach number
```

```python
_M = M.subs(subList)
subList.update({'M':_M})
print "Mach number = " + "%.2f" % _M #show N digits after decimal point
```

```python
_eta = eta.subs(subList)
subList.update({'eta':_eta})
_eta
```

2) Calculate the mechanical stream power using Equation 11.22:

```python
_W_m = W_m.subs(subList)
subList.update({'W_m':_W_m})
print "W_m = " + "%.0f" % _W_m + " W" #show N digits after decimal point
```

3) Calculate the overall acoustic power using Equation 11.21:

```python
_W_a = W_a.subs(subList)
subList.update({'W_a':_W_a})
print "W_a = " + "%.2f" % _W_a + " W" #show N digits after decimal point
```

4) Calculate the overall sound power level using Equation 11.24:

```python
_L_w = L_w.subs(subList)
subList.update({'L_w':_L_w})
print "L_w = " + "%.0f" % _L_w + " (dB re 10^12 W)" #show N digits after decimal point
```

5) Calculate the overall sound power level at 10m using Equation 11.25:

```python
_L_p = L_p.subs(subList)
subList.update({'L_p':_L_p})
_L_p
print "L_p = " + "%.0f" % _L_p + " (dB re 20 uPa)" #show N digits after decimal point
```
